```python
import math
import copy
import numpy as np

X = "X"
O = "O"
EMPTY = None

# Khởi tạo bàn cờ
def initial_state():
    return [[EMPTY]*3 for _ in range(3)]

# Xác định người chơi hiện tại
def player(board):
    x = sum(row.count(X) for row in board)
    o = sum(row.count(O) for row in board)
    return X if x == o else O

# Các nước đi hợp lệ
def actions(board):
    return [(i, j) for i in range(3) for j in range(3) if board[i][j] is EMPTY]

# Thực hiện nước đi
def result(board, action):
    i, j = action
    if board[i][j] is not EMPTY:
        raise Exception("Invalid move")
    new_board = copy.deepcopy(board)
    new_board[i][j] = player(board)
    return new_board

# Kiểm tra thắng
def winner(board):
    lines = []

    # hàng + cột
    for i in range(3):
        lines.append(board[i])
        lines.append([board[0][i], board[1][i], board[2][i]])

    # đường chéo
    lines.append([board[0][0], board[1][1], board[2][2]])
    lines.append([board[0][2], board[1][1], board[2][0]])

    for line in lines:
        if line[0] is not None and line.count(line[0]) == 3:
            return line[0]
    return None

# Kết thúc game
def terminal(board):
    return winner(board) is not None or all(cell is not EMPTY for row in board for cell in row)

# Hàm lượng giá
def utility(board):
    w = winner(board)
    if w == X: return 1
    if w == O: return -1
    return 0

# Minimax + Alpha-Beta
def max_value(board, alpha, beta):
    if terminal(board):
        return utility(board)
    v = -math.inf
    for act in actions(board):
        v = max(v, min_value(result(board, act), alpha, beta))
        alpha = max(alpha, v)
        if beta <= alpha:
            break
    return v

def min_value(board, alpha, beta):
    if terminal(board):
        return utility(board)
    v = math.inf
    for act in actions(board):
        v = min(v, max_value(result(board, act), alpha, beta))
        beta = min(beta, v)
        if beta <= alpha:
            break
    return v

# Chọn nước đi tốt nhất
def minimax(board):
    turn = player(board)
    best_move = None

    if turn == X:
        best_val = -math.inf
        for act in actions(board):
            val = min_value(result(board, act), -math.inf, math.inf)
            if val > best_val:
                best_val = val
                best_move = act
    else:
        best_val = math.inf
        for act in actions(board):
            val = max_value(result(board, act), -math.inf, math.inf)
            if val < best_val:
                best_val = val
                best_move = act

    return best_move

# In bàn cờ
def print_board(board):
    print(np.array(board))

# Main
if __name__ == "__main__":
    board = initial_state()
    user = input("Chọn X hoặc O: ").upper()
    ai = O if user == X else X

    while not terminal(board):
        print_board(board)
        turn = player(board)

        if turn == user:
            r = int(input("Row (0-2): "))
            c = int(input("Col (0-2): "))
            if (r, c) in actions(board):
                board = result(board, (r, c))
            else:
                print("Không hợp lệ!")
        else:
            print("AI đang đi...")
            move = minimax(board)
            board = result(board, move)

    print_board(board)
    w = winner(board)
    if w:
        print("Winner:", w)
    else:
        print("Hòa!")
```

```python
import math

# Khởi tạo bàn cờ 3x3
ban_co = [[' ' for _ in range(3)] for _ in range(3)]

def kiem_tra_thang_thua(b):
    for i in range(3):
        if b[i][0] == b[i][1] == b[i][2] and b[i][0] != ' ':
            return b[i][0]
        if b[0][i] == b[1][i] == b[2][i] and b[0][i] != ' ':
            return b[0][i]

    if b[0][0] == b[1][1] == b[2][2] and b[0][0] != ' ':
        return b[0][0]
    if b[0][2] == b[1][1] == b[2][0] and b[0][2] != ' ':
        return b[0][2]

    for hang in b:
        if ' ' in hang:
            return None
    return 'Hoa'

def thuat_toan_alpha_beta(b, do_sau, alpha, beta, la_max):
    kq = kiem_tra_thang_thua(b)
    if kq == 'O': return 1
    if kq == 'X': return -1
    if kq == 'Hoa': return 0

    if la_max:
        best = -math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'O'
                    val = thuat_toan_alpha_beta(b, do_sau+1, alpha, beta, False)
                    b[i][j] = ' '
                    best = max(best, val)
                    alpha = max(alpha, val)
                    if beta <= alpha:
                        return best
        return best
    else:
        best = math.inf
        for i in range(3):
            for j in range(3):
                if b[i][j] == ' ':
                    b[i][j] = 'X'
                    val = thuat_toan_alpha_beta(b, do_sau+1, alpha, beta, True)
                    b[i][j] = ' '
                    best = min(best, val)
                    beta = min(beta, val)
                    if beta <= alpha:
                        return best
        return best

def nuoc_di_cua_may():
    best = -math.inf
    move = (-1, -1)

    for i in range(3):
        for j in range(3):
            if ban_co[i][j] == ' ':
                ban_co[i][j] = 'O'
                val = thuat_toan_alpha_beta(ban_co, 0, -math.inf, math.inf, False)
                ban_co[i][j] = ' '
                if val > best:
                    best = val
                    move = (i, j)

    ban_co[move[0]][move[1]] = 'O'

def hien_thi_ban_co():
    print("\n-------------")
    for hang in ban_co:
        print(f"| {hang[0]} | {hang[1]} | {hang[2]} |")
        print("-------------")

# Main
print("TIC-TAC-TOE (ALPHA-BETA)")
while True:
    hien_thi_ban_co()
    kq = kiem_tra_thang_thua(ban_co)
    if kq:
        print("KẾT QUẢ:", kq)
        break

    # Người chơi
    while True:
        try:
            o = int(input("Chọn ô (1-9): ")) - 1
            h, c = divmod(o, 3)
            if 0 <= h < 3 and 0 <= c < 3 and ban_co[h][c] == ' ':
                ban_co[h][c] = 'X'
                break
            else:
                print("Ô không hợp lệ!")
        except:
            print("Nhập số 1-9!")

    if kiem_tra_thang_thua(ban_co):
        continue

    # AI
    print("Máy đang tính...")
    nuoc_di_cua_may()
```
