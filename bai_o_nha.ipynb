```python
import heapq

class NPuzzle:
    def __init__(self, start_state, n):
        self.n = n
        self.start_state = start_state
        self.goal_state = list(range(1, n*n)) + [0]

    def get_manhattan_distance(self, state):
        """Tính tổng khoảng cách Manhattan từ trạng thái hiện tại đến đích"""
        distance = 0
        for i, value in enumerate(state):
            if value != 0:
                # Vị trí hiện tại (hàng, cột)
                curr_row, curr_col = divmod(i, self.n)
                # Vị trí đích (hàng, cột)
                target_row, target_col = divmod(value - 1, self.n)
                distance += abs(curr_row - target_row) + abs(curr_col - target_col)
        return distance

    def is_solvable(self, state):
        """Kiểm tra xem cấu hình bài toán có thể giải được hay không"""
        inversions = 0
        state_no_zero = [x for x in state if x != 0]
        for i in range(len(state_no_zero)):
            for j in range(i + 1, len(state_no_zero)):
                if state_no_zero[i] > state_no_zero[j]:
                    inversions += 1

        # Với n lẻ: Số lượng cặp nghịch thế (inversions) phải là số chẵn
        if self.n % 2 != 0:
            return inversions % 2 == 0
        else:
            # Với n chẵn: Nghịch thế + vị trí hàng của ô trống (từ dưới lên) phải khớp tính chẵn lẻ
            blank_row_from_bottom = self.n - (state.index(0) // self.n)
            if blank_row_from_bottom % 2 == 0:
                return inversions % 2 != 0
            else:
                return inversions % 2 == 0

    def solve(self):
        if not self.is_solvable(self.start_state):
            return None, "Trạng thái này không có lời giải!"

        # (priority, g_cost, current_state, path)
        start_h = self.get_manhattan_distance(self.start_state)
        open_set = [(start_h, 0, self.start_state, [])]
        visited = {tuple(self.start_state): 0}

        print(f"Đang giải bài toán {self.n*self.n-1}-Puzzle...")

        while open_set:
            f, g, current_state, path = heapq.heappop(open_set)

            if current_state == self.goal_state:
                return path + [current_state], g

            blank_idx = current_state.index(0)
            r, c = divmod(blank_idx, self.n)

            # Các hướng di chuyển: Lên, Xuống, Trái, Phải
            for dr, dc in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                nr, nc = r + dr, c + dc

                if 0 <= nr < self.n and 0 <= nc < self.n:
                    neighbor_idx = nr * self.n + nc
                    new_state = list(current_state)
                    new_state[blank_idx], new_state
```

```python
import heapq

def tsp_a_star(dist_matrix):
    num_cities = len(dist_matrix)
    all_visited = (1 << num_cities) - 1

    # (priority, current_city, visited_mask, g_cost, path)
    start_node = (0, 0, 1 << 0, 0, [0])
    open_set = [start_node]

    def heuristic(current_city, visited_mask):
        # Ước lượng: Khoảng cách tới thành phố gần nhất chưa thăm
        min_dist = float('inf')
        found = False
        for i in range(num_cities):
            if not (visited_mask & (1 << i)):
                min_dist = min(min_dist, dist_matrix[current_city][i])
                found = True
        return min_dist if found else dist_matrix[current_city][0]

    while open_set:
        _, curr, mask, g, path = heapq.heappop(open_set)

        if mask == all_visited:
            return path + [0], g + dist_matrix[curr][0]

        for next_city in range(num_cities):
            if not (mask & (1 << next_city)):
                new_g = g + dist_matrix[curr][next_city]
                new_mask = mask | (1 << next_city)
                f = new_g + heuristic(next_city, new_mask)
                heapq.heappush(open_set, (f, next_city, new_mask, new_g, path + [next_city]))

# Ma trận khoảng cách giữa 4 thành phố
matrix = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]

path, cost = tsp_a_star(matrix)
print(f"Đường đi TSP: {path}")
print(f"Tổng chi phí: {cost}")
```
